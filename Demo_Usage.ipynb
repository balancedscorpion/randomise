### Simple Demo implementation of the Randomiser

```python
from app.randomise import Randomiser, HashAlgorithm, DistributionMethod
```

```python
randomiser = Randomiser(
    seed="experiment_1",
    proportions=[0.5, 0.5],
    table_size=10_000,
    algorithm=HashAlgorithm.XXH3,
    distribution=DistributionMethod.MAD 
)
```

```python
user_list = ["Alice","Alice","Alice","Alice","Alice","Alice","Alice","Alice","Alice","Bob","Alice"]
for user in user_list:
    print(user,randomiser.assign(user))
```
